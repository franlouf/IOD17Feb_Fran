````python
fig = plt.figure()
ax1 = fig.add_subplot(211)
np.random.seed(0)
x = stats.loggamma.rvs(5, size=500) + 5
# x follows a loggamma (i.e. non-bell-shaped) distribution - hence its probplot will not be linear at either end
# A Box-Cox transformation of x can achieve a more bell-shaped-looking distribution
# ANSWER

import scipy.stats as stats
from scipy import stats as sp_stats

stats.probplot(x, dist="norm", plot=ax1)
ax1.set_title('Original Data: Loggamma Distribution')

transformed_data, lambda_ = sp_stats.boxcox(x)
ax2 = fig.add_subplot(212)
stats.probplot(transformed_data, dist="norm", plot=ax2)
ax2.set_title(f'Box-Cox Transformed Data (lambda={lambda_:.2f})')

plt.show()
````

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, se

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m = np.mean(a)
  se = np.std(a, ddof=1) / np.sqrt(n)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))
````
