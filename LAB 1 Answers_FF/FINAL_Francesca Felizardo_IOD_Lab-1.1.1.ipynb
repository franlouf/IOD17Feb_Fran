## Working with lists

```python
# below is a list of 10 students (using just first name)
students = ['Alice','Ben','Cathy','Dawn','Emily','Fern','Emma','Sean','Felix','Hugh']
```

```python
# Print the list
print (students)
```

```python
# Print the list of students in the following format
# Index 01 	 Alice
# Index 02 	 Ben
# Index 03 	 Cathy
# ...
for i, student in enumerate(students, 1):
    print(f'Index {i:02d} \t {student}')
```

```python
# print the list of students in reverse order one student in a line with its index
# google search how to print list in reverse order
reverse_students = list(reversed(students))
for i, student in enumerate(reverse_students, 1):
    print(f'Index {i:02d} \t {student}')
```

```python
# print the student list in random order
# hint look up the function random.sample
import random
students_random = random.sample(students,10)
print(students_random)
```

```python
# list students in decreasing order of their exam marks given their marks below
# hint: look up the 'sorted' function
marks = [87,45,76,64,66,90,55,73,69,75]

students_and_marks = list(zip(students, marks))
sorted_students_and_marks = sorted(students_and_marks, key=lambda x: x[1], reverse=True)
sorted_students = [student for student, _ in sorted_students_and_marks]
print("Students in Decreasing Order of Marks:", sorted_students)
```

## Using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a set of x values
x = np.linspace(0, 10, 101)
# Print the inputs
print(x)
```

```python
# Create a figure using plt.figure
fig = plt.figure()
# Plot the sine of x
plt.plot(x, np.sin(x), linewidth=4)
# Plot the cosine of x
plt.plot(x, np.cos(x), '.')
plt.show()
```

## Plotting bar charts

```python
# Create two lists of 5 arbitrary numbers
array1 = [2, 7, 1, 9, 4]
array2 = [5, 3, 8, 6, 0]

# Plot two bar charts using plt.bar and the lists you've created above
# Label them "Example one" and "example two"

plt.bar(np.arange(2*len(array1))[::2]+1, array1, label="Example one", width=0.6)
plt.bar(np.arange(2*len(array1))[::2], array2, label="Example two", color='r', width=0.6)

plt.legend()
# Label x axis "Bar number"
plt.xlabel('Bar number')
# Label y axis "Bar height"
plt.ylabel('Bar height')
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code
# Create an array of a random Laplace distribution (a type of random number generator)
laplace_array = np.random.laplace(loc=15, scale=3, size=100)
# Create a histogram using plt.hist()
n, bins, patches = plt.hist(x=laplace_array, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(axis='both', alpha = 0.5)
plt.grid(axis='both', alpha =0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
plt.show()
```

```python
# Bonus: Create a scatter plot of random data
np.random.seed(1)

x = np.random.rand(100)*500
y = np.random.rand(100)*500

plt.scatter(x, y, color='red', alpha=0.5, edgecolors='black')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Scatter Plot")
plt.grid(True, alpha=0.5)
plt.show()
```

## Visualising the iris dataset with seaborn

```python
# Import seaborn library
import seaborn as sns
# Import load_iris
from sklearn.datasets import load_iris
# Load the dataset
dataset=load_iris()
# Place in pandas dataframe
data=pd.DataFrame(dataset["data"],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data["Species"]=dataset["target"]
data["Species"]=data["Species"].apply(lambda x: dataset["target_names"][x])
# Print top 10 rows
print(data.head(10))
```

```python
# Describe the dataframe
data.describe()
```

```python
# use seaborn pairplot function to plot relationship between variables
sns.pairplot(data, hue="Species");
```
